### fazendo duas planilhas para fazer o join depois

```python
planilha1=sp.groupby('city').agg(
    {'new_deaths': ['sum'], 'epidemiological_week': ['count'],'city_ibge_code': ['first'],
     'new_confirmed':['sum'],'city':['first'],'estimated_population_2019':['first']})
```

```python
planilha1.rename({'sum': 'new_deaths','count': 'epidemiological_week', 'first': 'city_ibge_code','sum':'new_confirmed','first': 'city' ,'first':'estimated_population_2019'}, axis=1)
```

```python
planilha1.head()
```

Fazendo a planilha repartindo desde o começo da vacinação

```python
antes_vacinacao = sp[sp['date']< '2021-01-17']
antes_vacinacao.shape
```

```python
depois_vacinacao = sp[sp['date']> '2021-01-17']
depois_vacinacao.shape
```
